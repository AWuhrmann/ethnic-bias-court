```python
import hashlib, math
import numpy as np
import pandas as pd
from collections import defaultdict
from scipy import stats
from inspect_ai.log import read_eval_log_samples

# BASELINE_PATH = "../logs/baseline/2026-04-24T10-44-26-00-00_tf-bench_hT6ntzc3ueePq2Y6sfJahZ.eval"
# BIAS_PATHS = {
#     "Sub-Saharan African": "../logs/bias_subsaharan_african/2026-04-24T10-55-53-00-00_bias-tf-bench_o2bWXPAMEhK5wy6ULAvs9r.eval",
#     "Arab Maghrebi":       "../logs/bias_arab_maghrebi/2026-04-24T11-06-49-00-00_bias-tf-bench_f6BoG5gtz2UjHCqrjDDnkU.eval",
#     "Swiss French":        "../logs/bias_swiss_french/2026-04-24T11-21-33-00-00_bias-tf-bench_W2ABUjQCooWmnAZhmEGSEM.eval",
#     "Albanian":            "../logs/bias_albanian/2026-04-24T11-32-31-00-00_bias-tf-bench_XmMQGhH9CQJ2FAtSQv9bD8.eval",
#     "Swiss German":        "../logs/bias_swiss_german/2026-04-24T11-43-49-00-00_bias-tf-bench_kCj3wYG8vtqTpozA3ExbPs.eval",
#     "Swiss Italian":       "../logs/bias_swiss_italian/2026-04-24T12-17-16-00-00_bias-tf-bench_kMEvkjjEekspQJbVW4xVM8.eval",
#     "South Asian":         "../logs/bias_south_asian/2026-04-24T12-26-41-00-00_bias-tf-bench_aGnme8J3pSxRZDyna32BaP.eval",
#     "Serbian":             "../logs/bias_serbian/2026-04-24T12-39-20-00-00_bias-tf-bench_hnJoAJiz8YNZmmW3xvKnwz.eval",
# }

# BASELINE_PATH = "../logs/baseline/2026-04-28T09-27-49-00-00_tf-bench_2YMGrjBuembzh5yu66iKEK.eval"
# BIAS_PATHS = {
#     "Albanian": "../logs/bias_albanian/2026-04-28T08-36-49-00-00_bias-tf-bench_neKQoMjJjaCe96ztYFG5Nm.eval",
#     "Arab Maghrebi": "../logs/bias_arab_maghrebi/2026-04-28T09-02-53-00-00_bias-tf-bench_HKCvmuDproEemLr9RtE8VL.eval",
#     "Serbian": "../logs/bias_serbian/2026-04-28T08-58-30-00-00_bias-tf-bench_HSAWotpdRNdVChgH6cDmCz.eval",
#     "South Asian": "../logs/bias_south_asian/2026-04-28T08-44-58-00-00_bias-tf-bench_9oRmtTvGg7oDRgxBramfsX.eval",
#     "Subsaharan African": "../logs/bias_subsaharan_african/2026-04-28T08-41-32-00-00_bias-tf-bench_ZpR7QkGaXYSwFyTLk8L6XX.eval",
#     "Swiss French": "../logs/bias_swiss_french/2026-04-28T08-49-22-00-00_bias-tf-bench_E9zkWx2q8epzLNq8Qb7vDZ.eval",
#     "Swiss German": "../logs/bias_swiss_german/2026-04-28T08-55-09-00-00_bias-tf-bench_8jzf2cByiEDVuW7oh5scny.eval",
#     "Swiss Italian": "../logs/bias_swiss_italian/2026-04-28T08-34-15-00-00_bias-tf-bench_PpMHNz2p3G4aJFUC7BycaS.eval",
#     "Turkish": "../logs/bias_turkish/2026-04-28T09-09-44-00-00_bias-tf-bench_4gSCq7urmN9ZhQrqsNRT33.eval",
# }

BASELINE_PATH = "../logs/baseline/2026-04-29T13-13-17-00-00_tf-bench_EFtMufeTiyywDGWNMA4bbu.eval"
BIAS_PATHS = {
    "Albanian": "../logs/bias_albanian/2026-04-29T13-22-30-00-00_bias-tf-bench_UufTciretnkKvXRzugvwRs.eval",
    "Arab Maghrebi": "../logs/bias_arab_maghrebi/2026-04-29T13-26-05-00-00_bias-tf-bench_BPfQz5THLvb92qwz7MrUuG.eval",
    "Serbian": "../logs/bias_serbian/2026-04-29T13-24-22-00-00_bias-tf-bench_QLstrDpdGvSqSwkMHBGCzh.eval",
    "South Asian": "../logs/bias_south_asian/2026-04-29T13-29-35-00-00_bias-tf-bench_n3AmNHm8BnA7zb7k5Tytvi.eval",
    "Subsaharan African": "../logs/bias_subsaharan_african/2026-04-29T13-27-49-00-00_bias-tf-bench_AMxRUVy2mrRWyKGwr5bjqm.eval",
    "Swiss French": "../logs/bias_swiss_french/2026-04-29T13-17-09-00-00_bias-tf-bench_QSvKTbEUs2fMCNkAozwJjD.eval",
    "Swiss German": "../logs/bias_swiss_german/2026-04-29T13-15-06-00-00_bias-tf-bench_2dW5kE7vwRKTtTWCbfsgQq.eval",
    "Swiss Italian": "../logs/bias_swiss_italian/2026-04-29T13-19-03-00-00_bias-tf-bench_jc4q48gkRN5tfvbHjEMqin.eval",
    "Turkish": "../logs/bias_turkish/2026-04-29T13-20-46-00-00_bias-tf-bench_hNHNJ4GRpF8uBCw8nfPBsp.eval",
}



def input_key(text):
    return hashlib.md5(text[:300].encode()).hexdigest()


def load_per_sample(path):
    """Returns DataFrame with one row per unique sample (mean over 3 epochs)."""
    by_key = defaultdict(list)
    for s in read_eval_log_samples(path):
        v = s.scores.get("tf_bench_llm_scorer")
        if not v or not isinstance(v.value, dict):
            continue
        refusal = v.value.get("refusal")
        quality = v.value.get("quality")
        if isinstance(refusal, float) and math.isnan(refusal):
            refusal = None
        inp = s.input if isinstance(s.input, str) else ""
        by_key[input_key(inp)].append(
            {
                "refusal": refusal,
                "quality": quality,
                "task": s.metadata.get("task"),
                "lang": s.metadata.get("lang"),
                "has_names": bool(s.metadata.get("substitutions")),
            }
        )
    rows = []
    for key, epochs in by_key.items():
        refs = [e["refusal"] for e in epochs if e["refusal"] is not None]
        quals = [e["quality"] for e in epochs if e["quality"] is not None and e["refusal"] == 0]
        rows.append(
            {
                "key": key,
                "refusal_rate": np.mean(refs) if refs else np.nan,
                "quality": np.mean(quals) if quals else np.nan,
                "task": epochs[0]["task"],
                "lang": epochs[0]["lang"],
                "has_names": epochs[0]["has_names"],
            }
        )
    return pd.DataFrame(rows).set_index("key")


print("Loading baseline...")
base_df = load_per_sample(BASELINE_PATH)
print(f"  {len(base_df)} samples")

bias_dfs = {}
for label, path in BIAS_PATHS.items():
    print(f"Loading {label}...")
    bias_dfs[label] = load_per_sample(path)
    print(f"  {len(bias_dfs[label])} samples")

print("Done.")
```

```python
# --- Build summary stats with bootstrap CIs ---

np.random.seed(42)
N_BOOT = 10_000

def bootstrap_ci(arr, n_boot=N_BOOT, ci=95):
    arr = arr[~np.isnan(arr)]
    if len(arr) == 0:
        return np.nan, np.nan
    boots = np.random.choice(arr, size=(n_boot, len(arr)), replace=True).mean(axis=1)
    lo = (100 - ci) / 2
    return np.percentile(boots, lo), np.percentile(boots, 100 - lo)

def paired_wilcoxon_p(base_arr, bias_arr):
    common = ~(np.isnan(base_arr) | np.isnan(bias_arr))
    b, s = base_arr[common], bias_arr[common]
    if len(b) < 10 or len(set(s - b)) <= 1:
        return np.nan
    _, p = stats.wilcoxon(s, b, alternative="two-sided")
    return p

rows = []

# Baseline row
ref = base_df["refusal_rate"].values
qual = base_df["quality"].dropna().values
rlo, rhi = bootstrap_ci(ref)
qlo, qhi = bootstrap_ci(qual)
rows.append({
    "ethnicity": "Baseline",
    "refusal_mean": np.nanmean(ref),
    "refusal_lo": rlo, "refusal_hi": rhi,
    "quality_mean": np.nanmean(qual),
    "quality_lo": qlo, "quality_hi": qhi,
    "delta_refusal": 0.0,
    "delta_lo": 0.0, "delta_hi": 0.0,
    "p_wilcoxon": np.nan,
    "n": len(base_df),
    "n_with_names": base_df["has_names"].sum(),
})

for label, df in bias_dfs.items():
    common_idx = base_df.index.intersection(df.index)
    b_ref = base_df.loc[common_idx, "refusal_rate"].values
    s_ref = df.loc[common_idx, "refusal_rate"].values
    b_qual = base_df.loc[common_idx, "quality"].values
    s_qual = df.loc[common_idx, "quality"].values

    diffs = s_ref - b_ref
    valid = ~np.isnan(diffs)
    dlo, dhi = bootstrap_ci(diffs[valid])
    rlo, rhi = bootstrap_ci(s_ref)
    qvals = s_qual[~np.isnan(s_qual)]
    qlo, qhi = bootstrap_ci(qvals)

    rows.append({
        "ethnicity": label,
        "refusal_mean": np.nanmean(s_ref),
        "refusal_lo": rlo, "refusal_hi": rhi,
        "quality_mean": np.nanmean(qvals),
        "quality_lo": qlo, "quality_hi": qhi,
        "delta_refusal": np.nanmean(diffs[valid]),
        "delta_lo": dlo, "delta_hi": dhi,
        "p_wilcoxon": paired_wilcoxon_p(b_ref, s_ref),
        "n": len(common_idx),
        "n_with_names": int(df.loc[common_idx, "has_names"].sum()),
    })

summary = pd.DataFrame(rows)

def sig_label(p):
    if np.isnan(p): return ""
    if p < 0.001: return "***"
    if p < 0.01:  return "**"
    if p < 0.05:  return "*"
    return "ns"

summary["sig"] = summary["p_wilcoxon"].apply(sig_label)

display_cols = ["ethnicity", "refusal_mean", "delta_refusal", "p_wilcoxon", "sig", "quality_mean", "n_with_names"]
summary[display_cols].round(4)
```

```python
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

# Sort by refusal mean for readability (baseline always first)
baseline_row = summary[summary.ethnicity == "Baseline"]
bias_rows = summary[summary.ethnicity != "Baseline"].sort_values("refusal_mean", ascending=False)
ordered = pd.concat([baseline_row, bias_rows], ignore_index=True)

PALETTE = px.colors.qualitative.Safe
colors = {
    row["ethnicity"]: ("#555555" if row["ethnicity"] == "Baseline" else PALETTE[i % len(PALETTE)])
    for i, row in ordered.iterrows()
}

# --- Fig 1: Refusal rate bar chart with 95% CI ---
fig1 = go.Figure()

for _, row in ordered.iterrows():
    eth = row["ethnicity"]
    err_lo = row["refusal_mean"] - row["refusal_lo"]
    err_hi = row["refusal_hi"] - row["refusal_mean"]
    label = f"{eth}<br><sub>{row['sig']}</sub>" if row["sig"] else eth
    fig1.add_trace(go.Bar(
        name=eth,
        x=[label],
        y=[row["refusal_mean"]],
        error_y=dict(type="data", array=[err_hi], arrayminus=[err_lo], visible=True, thickness=2, width=6),
        marker_color=colors[eth],
        hovertemplate=(
            f"<b>{eth}</b><br>"
            f"Refusal: {row['refusal_mean']:.3f}<br>"
            f"95% CI: [{row['refusal_lo']:.3f}, {row['refusal_hi']:.3f}]<br>"
            f"Wilcoxon p: {row['p_wilcoxon']:.4f} {row['sig']}<extra></extra>"
            if not np.isnan(row["p_wilcoxon"]) else
            f"<b>{eth}</b><br>Refusal: {row['refusal_mean']:.3f}<extra></extra>"
        ),
    ))

# Baseline reference line
base_refusal = baseline_row["refusal_mean"].values[0]
fig1.add_hline(
    y=base_refusal,
    line_dash="dash", line_color="#555555", line_width=1.5,
    annotation_text="baseline", annotation_position="right",
)

fig1.update_layout(
    title="Refusal Rate by Ethnicity (95% bootstrap CI, Wilcoxon vs baseline)",
    yaxis_title="Refusal rate (mean over 3 epochs)",
    xaxis_title="Ethnicity",
    showlegend=False,
    plot_bgcolor="white",
    yaxis=dict(gridcolor="#eeeeee", range=[0, 0.75]),
    font=dict(size=13),
    height=480,
)
fig1.show()
```

```python
# --- Fig 2: Delta refusal (vs baseline) with 95% CI ---
bias_only = ordered[ordered.ethnicity != "Baseline"].sort_values("delta_refusal", ascending=True)

fig2 = go.Figure()

for _, row in bias_only.iterrows():
    eth = row["ethnicity"]
    err_lo = row["delta_refusal"] - row["delta_lo"]
    err_hi = row["delta_hi"] - row["delta_refusal"]
    # Color: red if significant increase, blue if significant decrease, grey otherwise
    p = row["p_wilcoxon"]
    if not np.isnan(p) and p < 0.05:
        bar_color = "#d62728" if row["delta_refusal"] > 0 else "#1f77b4"
    else:
        bar_color = "#aaaaaa"

    fig2.add_trace(go.Bar(
        name=eth,
        x=[eth],
        y=[row["delta_refusal"]],
        error_y=dict(type="data", array=[err_hi], arrayminus=[err_lo], visible=True, thickness=2, width=6),
        marker_color=bar_color,
        hovertemplate=(
            f"<b>{eth}</b><br>"
            f"Δ refusal: {row['delta_refusal']:+.3f}<br>"
            f"95% CI: [{row['delta_lo']:+.3f}, {row['delta_hi']:+.3f}]<br>"
            f"Wilcoxon p: {row['p_wilcoxon']:.4f} {row['sig']}<extra></extra>"
        ),
    ))

fig2.add_hline(y=0, line_color="black", line_width=1)

# Significance annotations
for i, (_, row) in enumerate(bias_only.reset_index(drop=True).iterrows()):
    if row["sig"] not in ("", "ns"):
        fig2.add_annotation(
            x=row["ethnicity"],
            y=row["delta_hi"] + 0.01,
            text=row["sig"],
            showarrow=False,
            font=dict(size=14, color="black"),
        )

fig2.update_layout(
    title="Δ Refusal Rate vs Baseline (red=significant increase, blue=significant decrease, grey=ns)",
    yaxis_title="Δ refusal rate (bias − baseline)",
    xaxis_title="Ethnicity",
    showlegend=False,
    plot_bgcolor="white",
    yaxis=dict(gridcolor="#eeeeee", zeroline=True),
    font=dict(size=13),
    height=480,
)
fig2.show()
```

```python
# --- Fig 3: Box plots of per-sample refusal rate distributions ---

# Build long-form dataframe
records = []
for k, row in base_df.iterrows():
    records.append({"ethnicity": "Baseline", "refusal_rate": row["refusal_rate"], "quality": row["quality"]})

for label, df in bias_dfs.items():
    for k, row in df.iterrows():
        records.append({"ethnicity": label, "refusal_rate": row["refusal_rate"], "quality": row["quality"]})

long_df = pd.DataFrame(records)

# Order: baseline first, then sorted by median refusal
eth_order = ["Baseline"] + [
    e for e in long_df.groupby("ethnicity")["refusal_rate"].median()
    .drop("Baseline").sort_values(ascending=False).index
]

fig3 = go.Figure()
for eth in eth_order:
    vals = long_df.loc[long_df.ethnicity == eth, "refusal_rate"].dropna().values
    fig3.add_trace(go.Box(
        y=vals,
        name=eth,
        boxmean="sd",
        marker_color=colors.get(eth, "#888888"),
        line_width=1.5,
        hovertemplate=f"<b>{eth}</b><br>refusal_rate: %{{y:.3f}}<extra></extra>",
    ))

fig3.add_hline(
    y=base_df["refusal_rate"].mean(),
    line_dash="dash", line_color="#555555", line_width=1.5,
    annotation_text="baseline mean", annotation_position="right",
)

fig3.update_layout(
    title="Distribution of Per-Sample Refusal Rates (mean over 3 epochs per sample)",
    yaxis_title="Refusal rate",
    xaxis_title="Ethnicity",
    plot_bgcolor="white",
    yaxis=dict(gridcolor="#eeeeee"),
    showlegend=False,
    font=dict(size=13),
    height=500,
)
fig3.show()
```

```python
# --- Fig 5: Heatmap — Δ refusal rate by ethnicity × task/lang ---

dims = [("task", ["summarize", "translate"]), ("lang", ["de", "fr", "it", "en"])]

for dim, vals in dims:
    heat_data = []
    heat_text = []
    for label, df in bias_dfs.items():
        row_data, row_text = [], []
        common_idx = base_df.index.intersection(df.index)
        for v in vals:
            b_mask = base_df.loc[common_idx, dim] == v
            b_sub = base_df.loc[common_idx][b_mask]["refusal_rate"].dropna().values
            s_sub = df.loc[common_idx][b_mask]["refusal_rate"].dropna().values
            if len(b_sub) > 0:
                delta = np.mean(s_sub) - np.mean(b_sub)
                row_data.append(delta)
                row_text.append(f"{delta:+.3f}")
            else:
                row_data.append(np.nan)
                row_text.append("")
        heat_data.append(row_data)
        heat_text.append(row_text)

    eth_labels = list(bias_dfs.keys())
    fig_h = go.Figure(go.Heatmap(
        z=heat_data,
        x=vals,
        y=eth_labels,
        text=heat_text,
        texttemplate="%{text}",
        colorscale="RdBu_r",
        zmid=0,
        colorbar=dict(title="Δ refusal"),
        hovertemplate="<b>%{y}</b><br>%{x}: %{text}<extra></extra>",
    ))
    fig_h.update_layout(
        title=f"Δ Refusal Rate vs Baseline — by {dim}",
        xaxis_title=dim.capitalize(),
        yaxis_title="Ethnicity",
        font=dict(size=13),
        height=380,
        plot_bgcolor="white",
    )
    fig_h.show()
```
